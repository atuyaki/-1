```python
#ランダム関数の引用
import random
#playerに関する定義
class Player:
    def __init__(self,left,right,point):
        self.left = left
        self.right = right
        self.point = point
    def rplus(self,num,max):
        self.right = (num+self.right)%(max)
    def lplus(self,num, max):
        self.left = (num+self.left)%(max)
    def bunshin(self,num, max):
        self.rplus(num, max)
        self.lplus(num*(-1), max)
        self.point -= 1
    def addpt(self):
       addpoint=random.randint(0,5)
       if addpoint==0:
         self.point+=1
         print(self.name +"のポイントが加算されました！")
#子クラス
class Humanplayer(Player):
      def __init__(self,left,right,point,name):
        super().__init__(left,right,point)
        self.name = name

      def naming(self):
        self.name=input(self.name+"の名前を入力してください")
        print("あなたの名前は"+self.name+"です")
#モニター
def indi(a,b):
  print(a.name+"　左"+str(a.left)+"　右"+str(a.right)+"      ポイント: "+str(a.point)+"p")
  print(b.name+"　左"+str(b.left)+"　右"+str(b.right)+"      ポイント: "+str(b.point)+"p")
#動作
def turn(a,b):
  keynum=random.randint(5,8)
  print("今ターンのキーナンバーは"+str(keynum)+"です(^^)")
  #aのターン
  while True:
    sel1 = input(a.name+"の順番です。（1で自分の左手選択、2で右手選択、3で分身）")
    #左手で攻撃
    if sel1=="1":
      if a.left==0:
        print("自分の左手が0本なので無効")
        continue
      sel2 = input(a.name+"の順番です。（左手から、1で相手の左手攻撃、2で右手攻撃）")
      if sel2=="1":
        if b.left==0:
          print("相手の左手が0本なので無効")
          continue
        else:
          b.lplus(a.left,keynum)
          if b.left == 0:
            print(a.name+"がポイントをゲット！")
            a.point+=1
          break
      elif sel2=="2":
        if b.right==0:
          print("相手の右手が0本なので無効")
          continue
        else:
          b.rplus(a.left,keynum)
          if b.right == 0:
            print(a.name+"がポイントをゲット！")
            a.point+=1
          break
      else:
        print("そのコマンドは無効です")
        continue
    #右手で攻撃    
    elif sel1=="2":
      if a.right==0:
        print("自分の右手が0本なので無効")
        continue
      sel2 = input(a.name+"の順番です。（右手から、1で相手の左手攻撃、2で右手攻撃）")
      if sel2=="1":
        if b.left==0:
          print("相手の左手が0本なので無効")
          continue
        else:
          b.lplus(a.right, keynum)
          if b.left == 0:
            print(b.name+"がポイントをゲット！")
            a.point+=1          
          break
      elif sel2=="2":
        if b.right==0:
          print("相手の右手が0本なので無効")
          continue
        else:
          b.rplus(a.right, keynum)
          if b.right == 0:
            print(b.name+"がポイントをゲット！")
            a.point+=1          
          break
      else:
          print("そのコマンドは無効です")
          continue
    
    #分身
    elif sel1=="3":
        if a.point == 0:
          print("ポイントがありません")
        else:
          sel2 = input(a.name+"の順番です。（右手の本数を何本増減させるか半角数字で）")
          if int(sel2)>a.left or int(sel2)<a.right*(-1) or int(sel2)==a.left-a.right or int(sel2)==0:
            print("その数字は正しくありません")
            continue
          else:
            a.bunshin(int(sel2), keynum)
            break
    
    #どれでもない    
    else:
      print("そのコマンドは無効です")
      continue

#初期条件
p1 = Humanplayer(1,1,1,"player1(先攻)")
p2 = Humanplayer(1,1,1,"player2(後攻)")
p1.naming()
p2.naming()
while True:

  #ディスプレイ（ターン１）
  print("------------------------------------------")
  indi(p1,p2)
  print("------------------------------------------")
  if p1.left==0 and p1.right==0:
      print(p2.name+"の勝利！！！！！！！！")
      break
  
  #動作
  turn(p1,p2)
  #p1のポイント加算を決める
  p1.addpt()


  #ディスプレイ（ターン２）
  print("------------------------------------------")
  indi(p1,p2)
  print("------------------------------------------")

  if p2.left==0 and p2.right==0:
    print(p1.name+"の勝利！！！！！！！！")
    break
  
  #動作
  turn(p2,p1)
  #p2のポイント加算を決める
  p2.addpt()
print("finish")
input()
```
